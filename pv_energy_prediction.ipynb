```python
"""
This script performs data loading, cleaning, preprocessing, exploratory data analysis,
feature engineering, analysis, and visualization on photovoltaic data.
"""

import zipfile
import pandas as pd
import io
import seaborn as sns
import matplotlib.pyplot as plt

# --- Data Loading ---
# Define the path to the zip file
zip_file_path = '/content/Group_E_HS_Ansbach_Photovoltaic_Data.zip'

# Extract the CSV file from the zip archive and load it into a DataFrame
with zipfile.ZipFile(zip_file_path, 'r') as zf:
    file_list = zf.namelist()
    # Assuming the CSV file is the first file ending with .csv
    data_file_name = [f for f in file_list if f.endswith('.csv')][0]
    with zf.open(data_file_name) as f:
        # Read the CSV file, skipping the first few rows and using the row with meaningful names as header
        # Use delimiter ';' and engine 'python' as identified during EDA
        df = pd.read_csv(io.TextIOWrapper(f), delimiter=';', engine='python', skiprows=4, header=0)

# Drop the row containing units (which is now the first row after skipping)
df = df.drop(index=0)

print("Data loaded successfully.")
print("\nFirst few rows of the raw DataFrame:")
display(df.head())
print("\nDataFrame Info after loading:")
display(df.info())

# --- Data Cleaning and Preprocessing ---
# Convert 'Unit' to datetime and set as index
df['Unit'] = pd.to_datetime(df['Unit'], format='%d.%m.%Y %H:%M:%S')
df.set_index('Unit', inplace=True)

# Identify columns to convert to numeric (all object columns except the index)
cols_to_convert = df.select_dtypes(include='object').columns

# Convert identified columns to numeric after replacing comma with period
for col in cols_to_convert:
    df[col] = df[col].astype(str).str.replace(',', '.', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Handle missing values using forward fill (identified as a suitable method during cleaning)
df.fillna(method='ffill', inplace=True)

print("\nData cleaned and preprocessed successfully.")
print("\nFirst few rows of the preprocessed DataFrame:")
display(df.head())
print("\nDataFrame Info after preprocessing:")
display(df.info())
print("\nMissing values after preprocessing:")
display(df.isnull().sum())


# --- Exploratory Data Analysis (EDA) ---
print("\n--- Exploratory Data Analysis ---")
# Display descriptive statistics
print("\nDescriptive Statistics:")
display(df.describe())

# Generate and display correlation matrix heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()

# Time series plots for key columns
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(15, 20))

axes[0].plot(df.index, df['Globalstrahlungsfuehler Horizontal'])
axes[0].set_title('Global Horizontal Radiation over Time')
axes[0].set_ylabel('Radiation')

axes[1].plot(df.index, df['Elektro Aussentemperatur'])
axes[1].set_title('Outdoor Temperature over Time')
axes[1].set_ylabel('Temperature (°C)')

axes[2].plot(df.index, df['PV\xa0power generated 280'])
axes[2].set_title('PV Power Generated over Time')
axes[2].set_ylabel('Power (kW)')

axes[3].plot(df.index, df['NSHV (Haushaltstrom); consumption general electricity'])
axes[3].set_title('Household Electricity Consumption over Time')
axes[3].set_ylabel('Consumption (kW)')

plt.tight_layout()
plt.show()

# Histograms for selected features
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

axes[0, 0].hist(df['Globalstrahlungsfuehler Horizontal'], bins=50)
axes[0, 0].set_title('Distribution of Global Horizontal Radiation')
axes[0, 0].set_xlabel('Radiation')
axes[0, 0].set_ylabel('Frequency')

axes[0, 1].hist(df['Elektro Aussentemperatur'], bins=50)
axes[0, 1].set_title('Distribution of Outdoor Temperature')
axes[0, 1].set_xlabel('Temperature (°C)')
axes[0, 1].set_ylabel('Frequency')

axes[1, 0].hist(df['PV\xa0power generated 280'], bins=50)
axes[1, 0].set_title('Distribution of PV Power Generated')
axes[1, 0].set_xlabel('Power (kW)')
axes[1, 0].set_ylabel('Frequency')

axes[1, 1].hist(df['NSHV (Haushaltstrom); consumption general electricity'], bins=50)
axes[1, 1].set_title('Distribution of Household Electricity Consumption')
axes[1, 1].set_xlabel('Consumption (kW)')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# --- Feature Engineering ---
print("\n--- Feature Engineering ---")
# Extract time-based features from the index
df['hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek
df['day_of_year'] = df.index.dayofyear
df['month'] = df.index.month
df['year'] = df.index.year

print("\nTime-based features engineered successfully.")
display(df.head())
display(df.info())

# --- Analysis/Modeling ---
print("\n--- Analysis/Modeling ---")
# Select relevant columns for analysis
relevant_cols = ['PV\xa0power generated 280', 'Globalstrahlungsfuehler Horizontal', 'Elektro Aussentemperatur']
df_relevant = df[relevant_cols]

# Create scatter plots to visualize relationships
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

sns.scatterplot(data=df_relevant, x='Globalstrahlungsfuehler Horizontal', y='PV\xa0power generated 280', ax=axes[0])
axes[0].set_title('PV Power Generated vs. Horizontal Solar Radiation')
axes[0].set_xlabel('Global Horizontal Radiation')
axes[0].set_ylabel('PV Power Generated (kW)')

sns.scatterplot(data=df_relevant, x='Elektro Aussentemperatur', y='PV\xa0power generated 280', ax=axes[1])
axes[1].set_title('PV Power Generated vs. Outdoor Temperature')
axes[1].set_xlabel('Outdoor Temperature (°C)')
axes[1].set_ylabel('PV Power Generated (kW)')

plt.tight_layout()
plt.show()

# Calculate and print correlation coefficients
correlation_radiation = df_relevant['PV\xa0power generated 280'].corr(df_relevant['Globalstrahlungsfuehler Horizontal'])
correlation_temperature = df_relevant['PV\xa0power generated 280'].corr(df_relevant['Elektro Aussentemperatur'])

print(f"Correlation between PV Power Generated and Horizontal Solar Radiation: {correlation_radiation:.2f}")
print(f"Correlation between PV Power Generated and Outdoor Temperature: {correlation_temperature:.2f}")

# --- Visualization of PV Power vs. Time-based Features ---
print("\n--- Visualization of PV Power vs. Time-based Features ---")
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# PV Power vs. Hour (Box plot)
sns.boxplot(x='hour', y='PV\xa0power generated 280', data=df, ax=axes[0, 0])
axes[0, 0].set_title('PV Power Generated vs. Hour of Day')
axes[0, 0].set_xlabel('Hour of Day')
axes[0, 0].set_ylabel('PV Power Generated (kW)')

# PV Power vs. Day of Week (Box plot)
sns.boxplot(x='day_of_week', y='PV\xa0power generated 280', data=df, ax=axes[0, 1])
axes[0, 1].set_title('PV Power Generated vs. Day of Week')
axes[0, 1].set_xlabel('Day of Week (0=Monday, 6=Sunday)')
axes[0, 1].set_ylabel('PV Power Generated (kW)')

# PV Power vs. Month (Box plot)
sns.boxplot(x='month', y='PV\xa0power generated 280', data=df, ax=axes[1, 0])
axes[1, 0].set_title('PV Power Generated vs. Month')
axes[1, 0].set_xlabel('Month')
axes[1, 0].set_ylabel('PV Power Generated (kW)')

# PV Power vs. Year (Box plot)
sns.boxplot(x='year', y='PV\xa0power generated 280', data=df, ax=axes[1, 1])
axes[1, 1].set_title('PV Power Generated vs. Year')
axes[1, 1].set_xlabel('Year')
axes[1, 1].set_ylabel('PV Power Generated (kW)')

plt.tight_layout()
plt.show()

# PV Power vs. Day of Year (Scatter plot)
plt.figure(figsize=(15, 6))
sns.scatterplot(x='day_of_year', y='PV\xa0power generated 280', data=df, alpha=0.5, s=10)
plt.title('PV Power Generated vs. Day of Year')
plt.xlabel('Day of Year')
plt.ylabel('PV Power Generated (kW)')
plt.show()

print("\nScript execution completed.")
```
